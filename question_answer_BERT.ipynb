```python
import pdfplumber
pdf = pdfplumber.open('med1.pdf')
page = pdf.pages[0]
page1 = pdf.pages[1]
pdf_txt = page.extract_text() + page1.extract_text()
pdf.close()
```

```python
import torch
from transformers import BertForQuestionAnswering
model = BertForQuestionAnswering.from_pretrained('bert-large-uncased-whole-word-masking-finetuned-squad')
```

```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-large-uncased-whole-word-masking-finetuned-squad')
```

```python
answer_text = pdf_txt
question = "Who is Susan's doctor?"
```

```python
input_ids = tokenizer.encode(question, answer_text, max_length=512, truncation=True)
print (f'We have about {len(input_ids)} tokens generated')

tokens = tokenizer.convert_ids_to_tokens(input_ids)
print(" ")
print('Some examples of token-input_id pairs:')

for i, (token,inp_id) in enumerate(zip(tokens,input_ids)):
    print(token,":",inp_id)
```

```python
sep_idx = tokens.index('[SEP]')

# we will provide including [SEP] token which seperates question from context and 1 for rest.
token_type_ids = [0 for i in range(sep_idx+1)] + [1 for i in range(sep_idx+1,len(tokens))]
print(token_type_ids)
```

```python
out = model(torch.tensor([input_ids]), token_type_ids=torch.tensor([token_type_ids]))

start_logits,end_logits = out['start_logits'],out['end_logits']
print(start_logits)
```

```python
# Find the tokens with the highest `start` and `end` scores.
answer_start = torch.argmax(start_logits)
answer_end = torch.argmax(end_logits)
answer = ' '.join(tokens[answer_start:answer_end+1])
print('Answer: "' + answer + '"')
```
